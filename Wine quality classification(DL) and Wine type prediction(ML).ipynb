From the above we can see there are many outliers in the given dataset especially in the columns residual sugar, free sulfur dioxide, total sulfur dioxide.So, we remove them to improve model accuracy.

# Now lets train our quality prediction model using ML 

```python
#Splitting the dataset into train and test set
X=df2.drop('type',axis=1)
y=df2['type']
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
#Imoprting our desired model.
#Here I've used logistic regression as the wine type has binary value(either red or white) i.e.
#10 for white wine and 11 for red wine
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=42)
```

```python
#training our model
model.fit(X_train,y_train)
```

```python
#Getting accuracy of model
score=model.score(x_test,y_test)
print(f"Accuracy of logistic regression model is {score*100:.2f}%")
```

```python
#Predicting values
y_pred=model.predict(x_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

From y_test and y_pred we can see the model's prediction is pretty good!!

```python
#Checing model's metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
print(f"MAE of logistic regression model is {mean_absolute_error(y_test,y_pred):.2f}")
print(f"R2_score of logistic regression model is {r2_score(y_test,y_pred):.2f}")
```

We can see the mean absolute error is 0.02 (very low) and <br>the R2_score is 0.91 which is high.<br>Thus we can say we've got a pretty good model here with accuracy of 98%

# Now, lets create the model for quality prediction using neural networks.

```python
P=df1.drop('type',axis=1)
P
```

```python
#Getting train and test data
df_train=P.sample(frac=0.7,random_state=42)
df_valid=P.drop(df_train.index)
```

```python
df_train
```

```python
df_valid
```

```python
#Making the data into proper train and test sets to feed into the NN.
Xn_train = df_train.drop('quality', axis=1)
Xn_valid = df_valid.drop('quality', axis=1)
from keras.utils import to_categorical
yn_train = to_categorical(df_train['quality'])
yn_valid = to_categorical(df_valid['quality'])
```

```python
Xn_train.shape
```

```python
yn_train
```

# Creating our NN model using keras

```python
from tensorflow import keras
from tensorflow.keras import layers
from keras.optimizers import SGD,RMSprop,Adam

#defining our optimisers by defining a learning rate.
opt1=SGD(lr=0.01,momentum=0.9)
opt2=RMSprop(lr=0.01)
opt3=Adam(lr=0.01)

#Creating our model with 4 hidden layers and 1 output layer with each layer having 1024units
mdl=keras.Sequential([
    layers.BatchNormalization(input_shape=[11]),
    layers.Dense(1024, activation='relu'),
    layers.Dropout(rate=0.3),
    layers.Dense(1024, activation='relu'),
    layers.Dropout(rate=0.3),
    layers.BatchNormalization(),
    layers.Dense(1024, activation='relu'),
    layers.Dropout(rate=0.3),
    layers.BatchNormalization(),
    layers.Dense(1024, activation='relu'),
    layers.Dropout(rate=0.3),
    layers.BatchNormalization(),
    layers.Dense(3,activation='softmax'),
])

#Adding features to our model
mdl.compile(
     optimizer='adam',#or opt1 or opt2 or opt3.I've used  after testing as it gives best result.
    loss='categorical_crossentropy',#since it is the best classification loss evaluation metrics
    metrics=['accuracy']#since we're doing classification we need to check the model's accuracy
)

#Creating early stopping to stop training to prevent overfitting
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    min_delta=0.001, # minimium amount of change to count as an improvement
    patience=20, # how many epochs to wait before stopping
    restore_best_weights=True,
)
```

```python
#Fitting/Trainig our created model
new_history = mdl.fit(
    Xn_train, yn_train,
    validation_data=(Xn_valid, yn_valid),
    batch_size=100,
    epochs=500,
    callbacks=[early_stopping], # put your callbacks in a list
    verbose=0,  # turn off training log
)

new_history_df = pd.DataFrame(new_history.history)
new_history_df.loc[:, ['loss', 'val_loss']].plot();
new_history_df.loc[:, ['accuracy', 'val_accuracy']].plot();
print("Minimum validation loss: {:.3f}".format(new_history_df['val_loss'].min()))
print("Maximum validation accuracy: {:.3f}".format(new_history_df['val_accuracy'].max()))
```

```python
final_Xdata=P.drop('quality',axis=1)
final_ydata=P['quality']
final_ydata=to_categorical(final_ydata)
_, accuracy = mdl.evaluate(final_Xdata,final_ydata,batch_size=50, verbose=0)
print(f"The accuracy of our model is {accuracy*100 :.2f}%")
```

# Making predictions using our ML (typeprediction) and DL (qualityprediciton) model.

```python
def predictions(x):
    labels={"Low":0, "Medium":1, "High":2,'white':10,'red':11}
    quality_pred=np.argmax(mdl.predict(x),axis=-1)
    type_pred=model.predict(x)
    for k,v in labels.items():
        if v==quality_pred  :
            quality=k
        if type_pred==v:
            wine=k
    return [wine,quality]
```
